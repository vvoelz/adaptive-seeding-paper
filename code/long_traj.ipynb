```python
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import os, glob

trajfiles = glob.glob('/Volumes/My2TBPassportUltra/projects/long_mcsampler/nsteps1e9_init*.0_lam*.txt')
traj = []
for trajfile in trajfiles:
    traj.append( np.loadtxt(trajfile) )
```
